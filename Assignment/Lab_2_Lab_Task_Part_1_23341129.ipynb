```python
import numpy as np
from skimage import io, data, exposure
import matplotlib.pyplot as plt
```

# Task 1 (for bonus pts)

```python
# Import an image from skimage data
I = data.camera()

# Calculate the histogram of the image


def Hist(I):
  bins_hist =np.arange(0,256,1)
  hist = [0]*256
  for x in range(0, I.shape[0]):
    for y in range(0, I.shape[1]):
      i = I[x,y]                 
      hist[i] = hist[i]+1
  return hist, bins_hist

a=Hist(I)
plt.plot(a[1],a[0])
plt.xlabel("Intensity values")
plt.ylabel("# pixels")
plt.title("Histogram")
plt.show()
```

```python
hist,bins_hist=exposure.histogram(I,nbins=256,normalize=False)
plt.plot(bins_hist,hist)
plt.xlabel("Intensity values")
plt.ylabel("# pixels")
plt.title("Histogram")
plt.show()
```

# Task 2

```python
# Import the "good.jpeg" image
I = data.camera()

# normalize the image
I_norm = I/255.0

# perform arithmetic operations (contrast increase/decrease) on the image
I_add = np.clip(I_norm+0.5,0,1) # increase by 0.5
I_sub = np.clip(I_norm-0.4,0,1) # decrease by 0.4
I_mul = np.clip(I_norm*2,0,1) # multiply by 2
I_div = np.clip(I_norm/3,0,1) # divide by 3
```

# Task 3

```python
# plot in a (5, 2) grid 
plt.figure(figsize=(5,10))

plt.subplot(5,2,1)
io.imshow(I)
plt.title("1st")

plt.subplot(5,2,2)
plt.plot(bins_hist,hist)
plt.title("1st")

plt.subplot(5,2,3)
io.imshow(I_add)
plt.title("2nd")

h1,b1=exposure.histogram(I_add,nbins=256,normalize=False)
plt.subplot(5,2,4)
plt.plot(b1,h1)
plt.title("2nd")

plt.subplot(5,2,5)
io.imshow(I_sub)
plt.title("3rd")

h2,b2=exposure.histogram(I_sub,nbins=256,normalize=False)
plt.subplot(5,2,6)
plt.plot(b2,h2)
plt.title("3rd")

plt.subplot(5,2,7)
io.imshow(I_mul)
plt.title("4th")

h3,b3=exposure.histogram(I_mul,nbins=256,normalize=False)
plt.subplot(5,2,8)
plt.plot(b3,h3)
plt.title("4th")

plt.subplot(5,2,9)
io.imshow(I_div)
plt.title("5th")

h4,b4=exposure.histogram(I_div,nbins=256,normalize=False)
plt.subplot(5,2,10)
plt.plot(b4,h4)
plt.title("5th")
```
